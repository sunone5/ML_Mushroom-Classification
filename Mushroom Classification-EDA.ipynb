## Data pre-processing

### Identifying irrelevant features

Feature selection decisions were made based upon filtering methods. But before determining the level of influence of each feature, I wanted to find out which features were totally useless. To do this, two methods were used.

* Chi-Square hypothesis testing, on the data in it’s raw form (1 irrelevant feature found).

* Permutation hypothesis testing of each feature’s mean difference between the poisonous and edibles, after the data was converted into binary form (4 irrelevant features found).

#### Now a Chi-squared test will be run on all the columns to see if there are any features which are redundant across both classes.

```python
from scipy.stats import chi2_contingency,chi2
def chi_square_test(feature):
    p = list(df[df['class']=='p'][feature].value_counts())
    e = list(df[df['class']=='e'][feature].value_counts())
    if len(p)-len(e)>0:
        for k in range(len(p)-len(e)):
            e.append(0)
    if len(p)-len(e)<0:
        for j in range(len(e)-len(p)):
            p.append(0)
    obs = np.array([p,e])
    chi_square_stat, prob, dof, ex = chi2_contingency(obs, correction=False)
    value = chi2.ppf(0.99,dof)
    if (value<chi_square_stat) and (prob<0.01):
        return "Reject Null"
    else:
        return "Fail to reject"
    
drop_these_chi2 = []
for column in df.drop('class',axis=1).columns:
    if chi_square_test(column) == 'Fail to reject':
        drop_these_chi2.append(column)
df.drop(drop_these_chi2,axis=1,inplace=True)
print(drop_these_chi2)
```

* Only ['veil-type'] can be dropped. We failed to reject the null hypothesis with this column that the distribution of this variable is the same amongst both classes so can get rid of it.

* Now we are using permutation hypothesis testing on dummified variables. These will be in the form of T-tests.

## Data Encoding process
* As per observation the dataset has values in text format . machine learning model wouldn’t be able to process letters. 
* We need to convert all the unique categorical values to integers. 
* But we have to reduce the alphabetical relationship between each category 
* Thus we perform pandas encoding using pandas .get_dummies()

Convert categorical variable into dummy/indicator variables.

```python
from numpy import nan

def return_nan(x):
    if x=='?':
        return nan
    else:
        return x

df['stalk-root'] = df['stalk-root'].apply(lambda x:return_nan(x))

features = df.drop('class',axis=1).columns
df['class'].replace({'p':1,'e':0},inplace=True)
df = pd.concat([df.drop(features,axis=1),pd.get_dummies(df[features])],axis=1)
df = df.sort_values(by='class',ascending=True).reset_index().drop('index',axis=1)
class_0_max = max(df.loc[df['class']==0].index)+1

# Engineered Features
features = df.drop('class',axis=1).columns
print(df.shape)
df.head()
```

Using the pandas .get_dummies() function the table filled with entirely binary data, where 1 is present if a feature of a given column was present, and 0 otherwise.

After converting to binary format, the original 23 columns were transformed to 117 columns. No rows were dropped.

```python
# put engineered features into list
features = list(features)
features
```

The null hypothesis is that distribution of a given feature is the same between both the edible mushrooms and the poisonous ones. I will test every feature to check for redundant features.

The steps of the permutation hypothesis testing are as follows.

* Order the table by ascending order.
* Find the mean difference of the feature between class 0 and class 1, save it.
* Scramble the original table order.
* Find the mean difference .
* Repeat steps 3,4 many times (atleast 1000), appending each mean value. This creates a normal distribution
* Calculate the percentage of samples more extreme than the original mean difference - this is the p-value.
* If p>0.05 fail to reject null hypothesis, and consider the feature equally distributed between both edible and poisnous, and therefor redundant.

```python
df = df.sort_values(by='class',ascending=True).reset_index().drop('index',axis=1)
class_0_max = max(df.loc[df['class']==0].index)+1
class_0_max
```

```python
from sklearn.utils import shuffle

def mean_diff(data,column):
    mean_p = np.mean(data[data['class']==1][column])
    mean_e = np.mean(data[data['class']==0][column])
    return mean_e - mean_p

def permutate_mean_differences(data,column,number_of_permutations):
    means = []
    for i in range(number_of_permutations):
        permutation = shuffle(data[column],random_state=i)
        a = np.mean(permutation.loc[:class_0_max])
        b = np.mean(permutation.loc[class_0_max:])
        means.append(a-b)
    return pd.DataFrame(means,columns=['means'])

number_of_permutations = 1000
redundant_features = []

for column in features:
    original_mean_diff = mean_diff(df,column)
    _ = permutate_mean_differences(df,column,number_of_permutations)
    if original_mean_diff>0:
        p = np.sum(_['means'] >= original_mean_diff) / number_of_permutations
    elif original_mean_diff<0:
        p = np.sum(_['means'] <= original_mean_diff) / number_of_permutations
    else:
        redundant_features.append(column)
    if p>0.05:
        redundant_features.append(column)

redundant_features
```

```python
df.drop(redundant_features,axis=1,inplace=True)
print(df.shape)
df.head()
```
